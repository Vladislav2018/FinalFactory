```python
costs_pr2 = list(pr2['costs'])
```

```python
i,j = 0,0
while i < len(costs_pr2):
    base = costs_pr2[i]
    if i < len(costs_pr2)-1:
        j = i+1 
        while j < len(costs_pr2):
            delta = abs(base - costs_pr2[j])
            if delta < 1:
                costs_pr2.remove(costs_pr2[j])
            j+=1
    i+=1
```
